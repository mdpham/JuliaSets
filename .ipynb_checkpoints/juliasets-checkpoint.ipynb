```julia
# Make video (s=1024) at 2 framerate:
#   ffmpeg -f image2 -framerate 2 -i %06d.jpg -s 1024x1024 ./test.avi

# GENERATE JULIA SET IMAGES
using FileIO
using Images
using ImageView
using PyPlot

# Get [-2,2]x[-2,2] complex plane
#  s = number of pixels
function complexplane(s)
    x = linspace(-2,2,s)
    y = linspace(2,-2,s)
    Z = zeros(Complex,s,s)
    for i = 1:length(y)
        for j = 1:length(x)
            Z[i,j] = x[j] + y[i]*im
        end
    end
    return Z
end

function juliaf(c)
    function f(z)
        z.^2 + c 
    end
end

function juliaimage(c,s,)
    f = juliaf(c)
    Z = complexplane(s)
    # Start image at full intensity, decrement for each iteration
    itcount = zeros(size(Z)) + 255
    Tend = 255/5
    for t = 1:Tend
        Z = f(Z)
        subtwo = abs(Z) .< 2
        itcount[subtwo] = itcount[subtwo] - 5
    end
    itcount = itcount./255
    return itcount
end
```

```julia
c = 0.279
s = 1024
jimg = juliaimage(c,s)
# Plot histogram of how long each pixel took to diverge
# h = PyPlot.plt[:hist](jimg[:])
colorview(Gray, jimg)
colorview(Gray, 1-jimg)
```

```julia
# Vary complex component, fix real
tic()
s = 1024
fr = 100 # number of frames
count = 0
re = 0.3
for c = linspace(-1,1,fr)
    count = count+1
    jframe = juliaimage(re+c*im,s)
    filename = lpad(string(count),6,"0")
    save("temp/white/$filename.jpg", colorview(Gray,jframe))
    save("temp/black/$filename.jpg", colorview(Gray,1-jframe))
end
toc()
```
